Start with some fake data

```python
import numpy as np
data = np.array([0, 1, 0, -1, 0])
t = np.linspace(0, np.pi*2, 5)
```

```python
%matplotlib inline
import pylab
pylab.plot(t, data)
pylab.scatter(t, data)
pylab.show()
```

Here are two different curves that fit that data exactly

```python
t2 = np.linspace(0, np.pi*2, 100)
curve1 = np.sin(t2)
curve2 = np.sin(t2*5)
```

```python
pylab.plot(t2, curve1)
pylab.plot(t2, curve2)
pylab.scatter(t, data)
pylab.show()
```
